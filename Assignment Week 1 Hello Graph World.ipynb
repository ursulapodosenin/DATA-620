```python
# Installing the required packages
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Initializing an empty graph
G = nx.Graph()
```

```python
# Names
nodes = ["Alice", "Ben", "Clara", "Daniel", "Eva", "Felix", "Grace", "Hannah", "Ivan", "Julia"]
G.add_nodes_from(nodes)
```

```python
# Defining the edges following the Krackhardt kite structure
edges = [
    ("Alice", "Ben"),
    ("Alice", "Clara"),
    ("Ben", "Daniel"),
    ("Clara", "Daniel"),
    ("Clara", "Felix"),
    ("Daniel", "Eva"),
    ("Felix", "Grace"),
    ("Eva", "Grace"),
    ("Grace", "Hannah"),
    ("Hannah", "Ivan"),
    ("Hannah", "Julia"),
    ("Ivan", "Julia")
]
G.add_edges_from(edges)
```

```python
# Drawing the graph
plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G, seed=42)  # Force-directed layout
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1500, font_size=12, edge_color='black')
plt.title("Kite Graph", fontsize=16)
plt.show()
```
